```python
import pandas as pd

def clean_cpi_wide(file_path):
    df = pd.read_excel(file_path)
    df.columns = [
        "Year","Jan","Feb","Mar","Apr","May","Jun",
        "Jul","Aug","Sep","Oct","Nov","Dec"
    ]

    for m in ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]:
        df[m] = pd.to_numeric(df[m], errors="coerce")

    out = df.copy()

    for m in ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]:
        out[m] = df[m].pct_change() * 100

    return out

files = [
    "CPI Eggs Data 2010 - Sep. 2025.xlsx",
    "CPI Coffee Data 2010 - Sep. 2025.xlsx",
    "CPI Bread Data 2010 - Sep. 2025.xlsx"
]

for f in files:
    df = clean_cpi_wide(f)
    df.to_csv(f.replace(".xlsx","_inflation.csv"), index=False)
    print("saved:", f.replace(".xlsx","_inflation.csv"))
```
