```python
import pandas as pd
import numpy as np

def process_cpi_chained(path, outfile):
    df = pd.read_excel(path, skiprows=7, header=None)
    header = df.iloc[2].astype(str).str.strip()
    df.columns = header
    df = df.drop(index=[0,1,2]).reset_index(drop=True)
    df = df[df["Year"].apply(lambda x: str(x).isdigit())]
    df["Year"] = df["Year"].astype(int)
    months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
    df[months] = df[months].apply(pd.to_numeric, errors="coerce")
    long = df.melt(id_vars="Year", value_vars=months, var_name="Month", value_name="CPI")
    month_map = {"Jan":1,"Feb":2,"Mar":3,"Apr":4,"May":5,"Jun":6,"Jul":7,"Aug":8,"Sep":9,"Oct":10,"Nov":11,"Dec":12}
    long["MonthNum"] = long["Month"].map(month_map)
    long["Date"] = pd.to_datetime(dict(year=long["Year"], month=long["MonthNum"], day=1))
    long = long.sort_values("Date").reset_index(drop=True)
    long["Inflation"] = ((long["CPI"] - long["CPI"].shift(1)) / long["CPI"].shift(1)) * 100
    long.loc[0,"Inflation"] = np.nan
    long.to_csv(outfile, index=False)
    return long

eggs = process_cpi_chained("CPI Eggs Data 2010 - Sep. 2025.xlsx", "eggs_clean.csv")
coffee = process_cpi_chained("CPI Coffee Data 2010 - Sep. 2025.xlsx", "coffee_clean.csv")
bread = process_cpi_chained("CPI Bread Data 2010 - Sep. 2025.xlsx", "bread_clean.csv")

print("Eggs:\n", eggs.head(), "\n")
print("Coffee:\n", coffee.head(), "\n")
print("Bread:\n", bread.head(), "\n")
```
